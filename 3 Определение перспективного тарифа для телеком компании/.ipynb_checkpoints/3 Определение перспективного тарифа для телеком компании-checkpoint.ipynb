```python
def load_df(name_of_df, separator=',', ind=None):
    """Функция загружает датафрэйм и выводит основные данные по нему.
    Аргументы:
    name_of_df - путь к датафрэйму;
    separator - разделитель;
    ind - столбцы для использования в качестве номеров строк"""
    
    df = pd.read_csv(name_of_df, sep=separator, index_col=ind)
    print('Выведем первые 5 строк таблицы:')
    print(df.head())
    print()
    print('Получим общую информацию о таблице:')
    df.info()
    print()
    print(f'Пропущенные значения по переменным:')
    print(df.isnull().sum())
    return df
```

## Предобработка данных <a class="anchor" id="third-bullet"></a>

Подготовим данные к дальнейшему анализу, проанализируем таблицы и попробуем найти в них аномальные или некорректные значения и исправить. Напишем функции для автоматизации построения графиков и подсчета пропущенных значений.

```python
def boxplot(dataframe, clmn, without_outliers=False, rng=None):
    """Функция автоматизирует построение графика boxplot. 
    Для наглядности при without_outliers=True
    взяли не 1,5 межквартильных размаха, а 2.
    Аргументы:
    dataframe - таблица с данными;
    clmn - столбец, по которому будет строиться боксплот;
    without_outliers - нужно ли построить боксплот без учета выбросов;
    rng - диапазон значений"""

    if not rng and without_outliers:
        Q1 = dataframe[clmn].quantile(0.25) 
        Q3 = dataframe[clmn].quantile(0.75)
        IQR = Q3 - Q1
        a = Q1 - 2 * IQR
        b = Q3 + 2 * IQR
        if a == b:
            a = dataframe[clmn].median() - dataframe[clmn].std()
            b = dataframe[clmn].median() + dataframe[clmn].std()
        rng = (a, b)
        
    dataframe.boxplot(column=clmn, grid=True)
    
    name = clmn
    try:
        name = name.split('_')
        name = ' '.join(name)
        plt.ylabel(f"{name} values")
    except:
        plt.ylabel(f"{name} values")
    plt.ylim(rng)
    if without_outliers:
        plt.title(f"Boxplot of {name} without outliers")
    else:
        plt.title(f"Boxplot of {name}")
    plt.show()
```

```python
def hist(dataframe, clmn, bns, without_outliers=False, rng=None):
    """Функция автоматизирует построение гистограммы. 
    Для наглядности при without_outliers=True
    взяли не 1,5 межквартильных размаха, а 2.
    Аргументы:
    dataframe - таблица с данными;
    clmn - столбец, по которому будет строиться гистограмма;
    without_outliers - нужно ли построить боксплот без учета выбросов;
    rng - диапазон значений"""
    
    if not rng and without_outliers:
        Q1 = dataframe[clmn].quantile(0.25) 
        Q3 = dataframe[clmn].quantile(0.75)
        IQR = Q3 - Q1
        a = max(Q1 - 2 * IQR, dataframe[clmn].min())
        b = Q3 + 2 * IQR
        rng = (a, b)
            
    dataframe.plot(kind='hist',
                   y=clmn,
                   bins=bns,
                   legend=True,
                   range = rng,
                   grid=True)
    name = clmn
    try:
        name = name.split('_')
        name = ' '.join(name)
        plt.xlabel(f"{name} values")
    except:
        plt.xlabel(f"{name} values")
    if without_outliers:
        plt.title(f"Histogram of {name} without outliers")
    else:
        plt.title(f"Histogram of {name}")
    plt.show()
```

```python
def vizualization_with_without_outliers(dataframe, clmn, 
    bns_with_o=10, bns_without_o=10, rage_for_boxplot=None, 
    rage_for_hist=None, number_of_graph = 5):
    """Вывод нескольких графиков с учетом выбросов и без них.
    Переменная number_of_graph отвечает за то, какой 
    график из 4 выводить. Если указана цифра 5, то выводятся все графики"""
    
    if number_of_graph == 5:
        boxplot(dataframe, clmn, without_outliers=False,
                rng=rage_for_boxplot)
        boxplot(dataframe, clmn, without_outliers=True)
        hist(dataframe, clmn, bns_with_o, without_outliers=False,
             rng=rage_for_hist)
        hist(dataframe, clmn, bns_without_o, without_outliers=True)
    elif number_of_graph == 1:
        boxplot(dataframe, clmn, without_outliers=False,
                rng=rage_for_boxplot)
    elif number_of_graph == 2:
        boxplot(dataframe, clmn, without_outliers=True)
    elif number_of_graph == 3:
        hist(dataframe, clmn, bns_with_o,
             without_outliers=False, rng=rage_for_hist)
    elif number_of_graph == 4:
        hist(dataframe, clmn, bns_without_o,
             without_outliers=True)
```

```python
def is_null(list_of_df):
    """Функция считает количество пропущенных значений во всех переданных датафрэймах
    Аргументы:
    list_of_df - список датафрэймов"""
    
    for df in list_of_df:
        print(df.isnull().sum())
        print()
```

```python
def float_to_int(df, columns, ceil=False):
    """Функция переводит значения в тип данных int.
    Аргументы:
    df - таблица с данными
    columns - наименование столбца, значения которого нужно перевести в тип int
    ceil - необходимость округления значений вверх"""
    
    if ceil:
        for column in columns:
            df.loc[:, column] = df.loc[:, column] \
                .apply(lambda x: math.ceil(x)).astype('int')
    else:
        for column in columns:
            try:
                df.loc[:, column] = df.loc[:, column] \
                    .astype('int')
            except:
                try:
                    df.loc[:, column] = df.loc[:, column] \
                        .astype('Int64')
                except:
                    df.loc[:, column] = pd.to_numeric(df.loc[:, column] \
                        .round(), errors='coerce').astype('Int64')
```

В данном случае имеет смысл округлить значения вверх: в случае, если длительность разговора 0 секунд, то она таковой и останется (так мы можем однозначно идентифицировать пропущенные звонки). В случае, если звонок был принят, и его длительность больше 0, в таком случае произойдёт округление до 1 целой минуты. Нам удобно такое преобразование в том числе и потому, что оператор списывает стоимость за несколько секунд, как за минуту, всегда округляя в большую сторону.

Большая часть телефонных разговоров длится от 2 до 11 минут.

Мы видим, что имеется достаточно много нулевых значений (почти 20%). Вероятно, это пропущенные звонки. Существует несколько вариантов обработки значений. Например, мы может оставить нулевые значения, а при создании таблицы с количеством звонков задать условие, чтобы учитывались только те звонки, длительность которых больше 0. Альтернативный вариант - избавиться от нулевых значений. Это приведет к снижению количества звонков, сгруппированных по пользователям и месяцам, однако, мы получим лишь фактические звонки, где разговор действительно состоялся, и за который оператором списались средства.
Еще одним вариантом является создание столбца, который будет однозначно сигнализировать о том, состоялся ли разговор (была ли его длительность больше 0) или нет (звонок был пропущен). Это позволит сохранить нам исходные данные, поэтому воспользуемся этим вариантом.

На этапе анализа данных продолжительность телефонных разговоров будет сгруппирована по пользователям и по месяцам. Теоретически возможен вариант, когда мы получим, что пользователь за конкретный месяц проговорил 0 минут. Такие наблюдения могут сместить среднюю продолжительность звонков в меньшую сторону. Для того, чтобы этого избежать, мы будем задавать условие, согласно которому будут учитываться только ненулевые значения длительности звонков, сгруппированных по месяцам.

```python
def merge_dataframes(list_of_dataframes, method='outer'):
    """Функция объединяет в себе переданные ей датафрэймы.
    Аргументы:
    list_of_dataframes - список с датафрэймами;
    method - способ объединения (по умолчанию outer)"""
    
    data = list_of_dataframes[0]
    for dataframe in list_of_dataframes[1:]:
        data = data.merge(dataframe, how=method, sort=True)
    return data
```

Посчитаем помесячную выручку с каждого пользователя. Для этого необходимо вычесть бесплатный лимит из суммарного количества звонков, сообщений и интернет-трафика, а остаток умножить на значение из тарифного плана. Затем прибавим абонентскую плату, соответствующую тарифному плану.

Для каждого абонента произведем расчет платы сверх тарифного плана за отправку сообщений, в зависимости от его тарифа. Для этого напишем функцию, которая принимает на вход датафрэйм с данными и с тарифами, и рассчитывает для каждого пользователя абонентскую плату.

```python
def profit_per_user(dataframe, tariffs):
    """Функция принимает итоговый датафрэйм с пользователями и датафрэйм с
    тарифами, после чего добавляет в исходный датафрэйм данные о прибыли по 
    каждому абоненту.
    Аргументы:
    dataframe - датафрэйм сгруппированный по месяцам;
    tariffs - датафрэйм с тарифами"""
    
    for tariff in tariffs['tariff_name'].unique():
        messages_included = float(tariffs.loc[tariffs['tariff_name'] == 
                                             tariff, 'messages_included'])
        rub_per_message = float(tariffs.loc[tariffs['tariff_name'] == 
                                             tariff, 'rub_per_message'])

        dataframe.loc[dataframe['tariff'] == tariff, 'price_per_messages'] = \
            dataframe.loc[(dataframe['tariff'] == tariff) & \
        (~dataframe['number_of_messages_per_month'].isna()), 
                          'number_of_messages_per_month'] \
                .apply(lambda x: (
                    x - messages_included) * \
                        rub_per_message if x - \
                        messages_included > 0 else 0)
        

        mb_per_month_included = float(tariffs.loc[tariffs['tariff_name'] == 
                                             tariff, 'mb_per_month_included'])
        rub_per_gb = float(tariffs.loc[tariffs['tariff_name'] == 
                                             tariff, 'rub_per_gb'])

        dataframe.loc[dataframe['tariff'] == tariff, 'price_per_internet'] = \
            dataframe.loc[(dataframe['tariff'] == tariff) & \
                          (~dataframe['internet_traffic_per_month'].isna()), 
                          'internet_traffic_per_month'] \
                .apply(lambda x: math.ceil((x - \
                                    mb_per_month_included) / 1024)
                       * rub_per_gb if x - \
                        mb_per_month_included > 0 else 0)

        

        minutes_included = float(tariffs.loc[tariffs['tariff_name'] == 
                                             tariff, 'minutes_included'])
        rub_per_minute = float(tariffs.loc[tariffs['tariff_name'] == 
                                             tariff, 'rub_per_minute'])
            
        dataframe.loc[dataframe['tariff'] == tariff, 'price_per_calls'] = \
            dataframe.loc[(dataframe['tariff'] == tariff) & \
                          (~dataframe['duration'].isna()), 
                          'duration'] \
                .apply(lambda x: math.ceil(x - \
                                    minutes_included)
                       * rub_per_minute if x - \
                        minutes_included > 0 else 0)

        
        dataframe.loc[dataframe['price_per_messages'].isna(),
                     'price_per_messages'] = 0
        dataframe.loc[dataframe['price_per_internet'].isna(),
                     'price_per_internet'] = 0
        dataframe.loc[dataframe['price_per_calls'].isna(),
                     'price_per_calls'] = 0
        

        rub_monthly_fee = float(tariffs.loc[tariffs['tariff_name'] == 
                                             tariff, 'rub_monthly_fee'])

        dataframe.loc[dataframe['tariff'] == tariff,
                      'monthly_payment_by_user'] = \
        dataframe.loc[dataframe['tariff'] == tariff, 'price_per_messages'] + \
        dataframe.loc[dataframe['tariff'] == tariff, 'price_per_internet'] + \
        dataframe.loc[dataframe['tariff'] == tariff, 'price_per_calls'] + \
        rub_monthly_fee

    return dataframe
```

## Анализ полученных данных <a class="anchor" id="fourth-bullet"></a>

В этом разделе мы проанализируем полученные данные. Составим краткий план работы.

Исходя из выборки, опишем поведение клиентов оператора:
* проанализируем, сколько минут разговора, сколько сообщений и какой объём интернет-трафика требуется пользователям каждого тарифа в месяц;
* посчитаем описательные статистики (среднее значение, дисперсию, стандартное отклонение);
* построим гистограммы;
* опишем распределения.

Важно, чтобы мы не учитывали (отфильтровывали) те наблюдения, которые равны нулю по числу отправленных смс, трафику интернета или минут звонков, т.к. это сместит средние значения в меньшую сторону. Также имеются NaN значения, однако, при расчете описательных статистик, они по умолчанию пропускаются, поэтому опасаться их не стоит.

Напишем функцию для описательных статистик. Нам необходимо посчитать средние значения, дисперсию и стандартное отклонение. Также добавим значения квартилей.

```python
def describe_stats(dataframe, list_of_features, by='tariff'):
    """Функция выводит основные описательные статистики.
    Аргументы:
    dataframe - таблица с данными;
    list_of_features - фичи, по которым необходимо рассчитать описательные статистики;
    by - по каким группам произвести разделение датафрэйма"""

    for feature in list_of_features:
        # group 1
        describe_1 = dataframe.loc[(dataframe[by] == dataframe[by] \
                                .unique()[0]) &
                                (dataframe[feature] != 0)][[feature]] \
                                .describe()
        variance_1 = dataframe.loc[(dataframe[by] == dataframe[by] \
                                    .unique()[0]) &
                                   (dataframe[feature] != 0)][[feature]].var()
        variance_1 = pd.DataFrame(np.array(variance_1), index=['var'], 
                                  columns=[feature])
        print(f"""Описательные статистики для пользователей тарифа 
            {dataframe[by].unique()[0]} по переменной {feature}""")
        display(pd.concat([describe_1, variance_1]))

        # group 2
        describe_2 = dataframe.loc[(dataframe[by] == dataframe[by] \
                                    .unique()[1]) &
                                   (dataframe[feature] != 0)][[feature]]. \
                                    describe()
        variance_2 = dataframe.loc[(dataframe[by] == dataframe[by] \
                                    .unique()[1]) &
                                   (dataframe[feature] != 0)][[feature]].var()
        variance_2 = pd.DataFrame(np.array(variance_2), index=['var'], 
                                  columns=[feature])
        print(f"""Описательные статистики для пользователей тарифа 
              {dataframe[by].unique()[1]} по переменной {feature}""")
        display(pd.concat([describe_2, variance_2]))
```

```python
describe_stats(df_tariffs_by_user, ['number_of_messages_per_month', 'internet_traffic_per_month', 'duration'])
```

Что примечательно, в среднем пользователи тарифа smart расходуют больше интернет-трафика, чем входит в стоимость тарифа, однако укладываются в длительность звонков и число сообщений. Дисперсия — метрика разброса. Это не просто среднее расстояние между значениями датасета и средним, а средний квадрат этого расстояния. Дисперсия для пользователей тарифа ultra по числу отправленных смс, расходу интернет-трафика, а также минутам разговора соответственно равна: 2110; 98 746 499; 93 338. А для пользователей тарифа smart дисперсия соответственно равна: 718; 34 340 092; 35 718.

У дисперсии есть один небольшой недостаток: единица её измерения — это квадрат исходной величины. Чтобы вернуться к исходной единице измерения, из дисперсии извлекают квадратный корень. Получившаяся величина называется стандартным отклонением. Стандартное отклонение для пользователей тарифа ultra соответственно равно: 46; 9937; 306. А для пользователей тарифа smart: 27; 5860; 189.

Дисперсия для пользователей тарифа smart в среднем в 4 раза меньше, чем для пользователей тарифа ultra. А стандартное отклонение, исходя из формулы (корень из дисперсии) - меньше в 2 раза.

 Напишем функцию с учетом фильтра нулевых значений, которая будет демонстрировать поведение пользователей 2 разных тарифов с помощью графиков.

```python
def hist_compare_and_describe(dataframe, list_of_features, by, ind, aggfunc='median', bns=25):
    """Функция принимает на вход dataframe а также список фичей, после чего разделяет их на группы за счет аргумента by.
    На основании разделения функция строит гистограммы для сравнения двух групп, а также выводит их описательные статистики.
    Аргументы:
    dataframe - таблица с данными;
    list_of_features - список с наименованиями фичей;
    by - наименование номинативного столбца, по кторому происходит разделение;
    ind - наименование столбца, который служет индексом в pivot_table;
    aggfunc - применяемая функция к элементам pivot_table (по умолчанию median);
    bns - определяет число корзин для гистограмм"""
    
    for feature in list_of_features:
        graph1 = dataframe.loc[(dataframe[by] == dataframe[by] \
                                .unique()[0]) & \
                               (dataframe[feature] != 0)] \
        .pivot_table(index=ind, values=feature, aggfunc=aggfunc)
        graph2 = dataframe.loc[(dataframe[by] == dataframe[by] \
                                .unique()[1]) & \
                               (dataframe[feature] != 0)] \
        .pivot_table(index=ind, values=feature, aggfunc=aggfunc)     
        
        ax = graph1.plot(kind='hist', y=feature, histtype='step', bins=bns, 
                         linewidth=3, alpha=0.7, 
                         label=dataframe[by].unique()[0])      
        graph2.plot(kind='hist', y=feature, histtype='step', 
                    bins=bns, linewidth=3, alpha=0.7,
                    label=dataframe[by].unique()[1], ax=ax, grid=True,
                    legend=True, figsize=(12, 5))

        name = feature
        try:
            name = name.split('_')
            name = ' '.join(name)
            plt.xlabel(f"{name} values")
        except:
            plt.xlabel(f"{name} values")
        plt.xlabel(f"{name} values")
        plt.title(f"Histogram of {name}")
        plt.show()
```

Посмотрим, сколько в среднем каждый из пользователей тарифов тратят минут на разговоры, сколько мегабайт интернет-трафика и сколько смс. Для этого воспользуемся функцией и визуализируем данные.

```python
hist_compare_and_describe(df_tariffs_by_user, ['number_of_messages_per_month', 'internet_traffic_per_month', 'duration'], 'tariff', 'user_id', aggfunc='median', bns=20)
```

Пользователей тарифа smart ожидаемо больше, поэтому по оси Y наибольшие значения достигаются за счет пользователей с тарифом smart. Однако, по оси X максимальные значения достигаются пользователями тарифа ultra: они расходуют больше интернет-трафика, отправляют больше смс и продолжительность их звонков тоже больше (хотя аналитически показать так это или нет мы сможем дальше, при помощи t-распределения, которое позволяет сравнить две выборки по средним значениям). 

Мы видим, что на первом графике (number_of_messages_per_month) распределение по пользователем тарифа smart и ultra ассиметрично и скошено вправо. В остальных случаях оно менее ассиметрично. Проверим распределения на нормальность, для этого воспользуемся тестом Шапиро-Уилка. Нулевой гипотезой будет то, что распределение является нормальным, а границей для принятия или отвержения нулевой гипотезы зададим 0.05.

```python
def is_normal(dataframe, list_of_features, by, tariff, alpha=0.05):
    """Функция рассчитывает вероятность того, что распределение является нормальным
    на основе теста Шапиро-Уилка, а также заданного значения alpha по каждой фиче.
    Аргументы:
    dataframe - таблица с данными;
    list_of_features - список с наименованиями фичей;
    by - наименование номинативного столбца, по которому происходит разделение;
    tariff - наименования тарифа
    alpha - граница для принятия или отвержения нулевой гипотезы"""

    for feature in list_of_features:
        data = dataframe.loc[(dataframe[by] == tariff) & \
                             (dataframe[feature] != 0) & \
                             (dataframe[feature].notna())][[feature]]
        stat, p = shapiro(data)
        print('Statistics=%.3f, p=%.3f' % (stat, p))
        if p > alpha:
            print(f'Распределение статистики {feature} похоже на нормальное (не отклоняем H0)', end='\n\n')
        else:
            print(f'Распределение статистики {feature} не похоже на нормальное (отклоняем H0)', end='\n\n')
```

```python
is_normal(df_tariffs_by_user, ['number_of_messages_per_month',
                               'internet_traffic_per_month', 'duration'], 'tariff', 'ultra')
```

```python
is_normal(df_tariffs_by_user, ['number_of_messages_per_month',
                               'internet_traffic_per_month', 'duration'],
          'tariff', 'smart')
```

Мы видим, что ни одно распределение нельзя отнести к нормальному.

## Проверка статистических гипотез <a class="anchor" id="fifth-bullet"></a>

Нам необходимо проверить статистические гипотезы:
* средняя выручка пользователей тарифов «Ультра» и «Смарт» различается;
* средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов.

Начнем с первой гипотезы. Сформулируем нулевую гипотезу (H0). H0 всегда формулируется так, чтобы использовать знак равенства. Другими словами, чтобы все значения, полученные эмпирически, не были экстремальными, а лежали там, где ожидаем их найти.

Нулевая гипотеза (H0) в данном случае будет говорить о том, что средние выручки пользователей тарифов «Ультра» и «Смарт» не различаются. В таком случае, альтернативной гипотезой (H1) будет являться гипотеза о том, что средние выручки пользователей тарифов «Ультра» и «Смарт» различаются. Такая альтернативная гипотеза называется двухсторонней. Её характерный признак — знак «не равно» в формулировке.

Есть смысл решать вопрос о принятии или отвержении нулевой гипотезы по p-value. Это вероятность получить наблюдаемый или ещё более далёкий от предполагаемого результат при условии, что нулевая гипотеза верна. Если это значение больше 10%, то нулевую гипотезу точно не стоит отвергать. Меньше — возможно, есть основания отвергнуть нулевую гипотезу. Общепринятые пороговые значения — 5% и 1%. В данном случае, примем p-value в размере 5%.

Для проверки гипотезы мы будем использовать t-критерий Стьюдента. t-критерий Стьюдента — общее название для статистических тестов, в которых статистика критерия имеет распределение Стьюдента. Наиболее часто t-критерии применяются для проверки равенства средних значений в двух выборках. Все разновидности критерия Стьюдента являются параметрическими и основаны на дополнительном предположении о нормальности выборки данных.

Однако, согласно центральной предельной теореме (ЦПТ), мы можем быть уверены, что  выборочное распределение выборочного среднего из любой генеральной совокупности распределено нормально вокруг среднего этой генеральной совокупности.

Определим, являются ли выборки независимыми или же зависимыми. В случае парных выборок имеются пары наблюдений (измерений) одного и того же объекта. Два раза измеряется одно и то же. В случае независимых выборок каждое наблюдение соответствует отдельному объекту, т.е. измеряются разные объекты. В нашем случае наши выборки независимы, поэтому воспользуемся методом scipy.stats.ttest_ind().

```python
def is_different_by_ttest(dataframe, list_of_features, by='tariff', alpha=0.05, eq_var=True):
    """Функция выявляет отличия в средних каждой 
    из двух выборок на основе t-критерия Стьюдента
    Аргументы:
    dataframe - таблица с данными;
    list_of_features - список фичей, по которым требуется выполнить анализ;
    by - столбец, по которому необходимо выполнить разбиение на 2 выборки
    alpha - граница для принятия или отвержения нулевой гипотезы"""
    
    
     
    for feature in list_of_features:
        var1 = dataframe.loc[(dataframe[by] == dataframe[by] \
                                 .unique()[0]) & \
                                (dataframe[feature] != 0)][[feature]].var()
        var2 = dataframe.loc[(dataframe[by] == dataframe[by] \
                                 .unique()[1]) & \
                                (dataframe[feature] != 0)][[feature]].var()
        if float(abs((var1 - var2) / var2)) > 0.05:
            eq_var = False
        tariff1 = dataframe.loc[(dataframe[by] == dataframe[by] \
                                 .unique()[0]) & \
                                (dataframe[feature] != 0)][[feature]]
        tariff2 = dataframe.loc[(dataframe[by] == dataframe[by] \
                                 .unique()[1]) & \
                                (dataframe[feature] != 0)][[feature]]
    
        results = st.ttest_ind(tariff1, tariff2, equal_var=eq_var)
        print('p-значение:', results.pvalue)
        if (results.pvalue < alpha):
            print("Отвергаем нулевую гипотезу, принимаем альтернативную")
        else:
            print("Не получилось отвергнуть нулевую гипотезу") 
```

```python
is_different_by_ttest(df_tariffs_by_user, ['monthly_payment_by_user'])
```

Мы получили p-value практически равное нулю. Это означает, что мы получили статистически значимое отличие двух средних выборок, и отвергаем нулевую гипотезу об их равенстве. То есть, вероятность получить такое или еще большее отклонение практически равна нулю.

Теперь проверим гипотезу о том, что средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов. Нулевой гипотезой (H0) в данном случае будет гипотеза о том, что средняя выручка пользователей из Москвы ничем не отличается от выручки пользователей из других регионов. Альтернативной гипотезой (H1). P-value также примем в размере 5%.

В таком случае можно однозначно порекомендовать отделу маркетинга уделять внимание тарифу "Smart", т.к. он обеспечивает большую часть прибыли. Возможно, анализ конкурентов с аналогичными тарифами и особые условия на первоначальных этапах по данному тарифу принесут хороший результат - увеличение прибыли компании.

Полученное значение p-value говорит о том, что хотя средний доход, приносимый пользователем из Москвы и всех остальных пользователей, неодинаков, с вероятностью в почти 51% такое или большее различие можно получить случайно. Это явно слишком большая вероятность, чтобы делать вывод о значимом различии между средними чеками.
