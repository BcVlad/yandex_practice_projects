```python
def load_df(name_of_df, separator=",", ind=None, parse_d=False):
    """Функция загружает датафрэйм и выводит основные данные по нему.
    Аргументы:
    name_of_df - путь к датафрэйму;
    separator - разделитель;
    ind - столбцы для использования в качестве номеров строк"""

    # Считаем и выведем первые 5 строк таблицы
    df = pd.read_csv(name_of_df, sep=separator, index_col=ind, parse_dates=parse_d)
    print("Выведем первые 5 строк таблицы:", end="\n\n")
    display(df.head())

    # Вывод основных описательных статистик, общей информации о таблице,
    # пропущенных значений, корреляции признаков, количества дубликатов в данных
    print(
        "\n\n\n\nОтобразим основные описательные статистики по признакам:", end="\n\n"
    )
    display(df.describe())
    print("\n\n\n\nПолучим общую информацию о таблице:", end="\n\n")
    display(df.info())
    print(f"\n\n\n\nПропущенные значения по признакам:", end="\n\n")
    display(pd.DataFrame(df.isnull().sum(), columns=["missing_values"]))
    print(f"\n\n\n\nКорреляция признаков:", end="\n\n")
    display(df.corr())
    return df, df.describe()
```

Проанализируем распределение количества заказов такси с помощью гистограммы, а также обратим внимание на основные описательные статистики.

```python
def hist(dataframe, clmn, bns, without_outliers=False, rng=None):
    """Функция автоматизирует построение гистограммы"""

    if not rng and without_outliers:
        Q1 = dataframe[clmn].quantile(0.25)
        Q3 = dataframe[clmn].quantile(0.75)
        IQR = Q3 - Q1
        a = max(Q1 - 2 * IQR, dataframe[clmn].min())
        b = Q3 + 2 * IQR
        rng = (a, b)

    dataframe.plot(kind="hist", y=clmn, bins=bns, legend=True, range=rng, grid=True)
    name = clmn
    try:
        name = name.split("_")
        name = " ".join(name)
        plt.xlabel(f"{name} values")
    except:
        plt.xlabel(f"{name} values")
    if without_outliers:
        plt.title(f"Histogram of {name} without outliers")
    else:
        plt.title(f"Histogram of {name}")

    plt.show()
```

Для анализа временного ряда построим также график, учитывающий автокорреляцию и частную автокорреляцию. Обычно эти графики используются для подбора параметров моделей семейства ARIMA.

```python
def time_s_plot(y, lags=None, figsize=(12, 7), style="bmh"):
    """Функция строит графики для анализа временных рядов:
    в том числе график автокорреляции и частной автокорреляции"""

    if not isinstance(y, pd.Series):
        y = pd.Series(y)
    with plt.style.context(style):
        fig = plt.figure(figsize=figsize)
        layout = (2, 2)
        ts_ax = plt.subplot2grid(layout, (0, 0), colspan=2)
        acf_ax = plt.subplot2grid(layout, (1, 0))
        pacf_ax = plt.subplot2grid(layout, (1, 1))

        y.plot(ax=ts_ax)
        ts_ax.set_title("Графический анализ временных рядов")
        sm.graphics.tsaplots.plot_acf(y, lags=lags, ax=acf_ax, alpha=0.5)
        sm.graphics.tsaplots.plot_pacf(y, lags=lags, ax=pacf_ax, alpha=0.5)

        print(f"Критерий Дики-Фуллера: p={round(sm.tsa.stattools.adfuller(y)[1], 5)}")

        plt.tight_layout()
    return
```

```python
time_s_plot(taxi_df["num_orders"], lags=35)
```

Критерий Дики-Фуллера отвергает нулевую гипотезу о нестационарности, но автокорреляционная функция всё ещё выглядит неидеально из-за большого числа значимых лагов. Используем преобразование Бокса-Кокса. Это преобразование используется в случае, если данные не проходят тест на нормальность, а о их распределении мало что известно.

```python
taxi_df["boxcox_transformed"] = boxcox1p(taxi_df.num_orders, 0.001)
```

```python
time_s_plot(taxi_df["boxcox_transformed"], lags=35)
```

Мы еще сумели уменьшить значение критерия Дики-Фуллера, однако, попробуем еще уменьшить значение p-value. В качестве следующих преобразований мы можем использовать сезонные разности методом .shift():

```python
taxi_df["box_season_diff"] = taxi_df["boxcox_transformed"] - taxi_df[
    "boxcox_transformed"
].shift(1, fill_value=0)
```

```python
time_s_plot(taxi_df["box_season_diff"], lags=35)
```

Теперь мы получили стационарный ряд: остатки стационарны, явных автокорреляций нет. По такому ряду можно строить модели и делать предсказания.

## Вывод

Нам удалось избавиться от нестационарности ряда с помощью преобразования Бокса-Кокса и использования сезонных разностей: p-value критерия Дики-Фуллера равно 0. Далее построим модели и спрогнозируем значения временного ряда.

# Обучение и тестирование моделей

Для предсказания значений временного ряда будем использовать несколько моделей:
* Линейная регрессия;

Начнём с линейной регрессии. Мы знаем, что ее главным преимуществом являются скорость обучения и небольшое число параметров для настройки. Однако, в случае недостаточного числа фичей точность предсказания будет не очень высока. Чтобы это предотвратить, добавим в наш датасэт новые фичи.

В качестве новых фичей можно использовать:
* час;
* день недели;
* является ли день выходным днем (принимает значения 0 или 1);
* лаг с соответствующим значением;
* скользящее среднее.

```python
def add_features(dataframe, max_lag, rolling_mean_size):
    """Функция создает дополнительные признаки,
    расширяя размерность признакового пространства"""

    data = pd.DataFrame(dataframe)
    data.columns = ["y"]
    data["hour"] = data.index.hour
    data["weekday"] = data.index.weekday
    #data["is_weekend"] = data.weekday.isin([5, 6]) * 1

    for lag in range(1, max_lag + 1):
        data[f"lag_{lag}"] = data["y"].shift(lag)

    data["rolling_mean"] = data["y"].shift().rolling(rolling_mean_size).mean()
    return data
```

```python
data_boxcox = add_features(taxi_df["boxcox_transformed"], 194, 39)
```

```python
data_boxcox_season_diff = add_features(taxi_df["box_season_diff"], 194, 39)
```

Напишем функции для построения графика, а также обучения модели. 

Стоит обратить внимание, что при обучении модели мы не можем подбирать гиперпараметры с помощью GridSearchCV, т.к. кросс-валидация способствует лику данных из "будущего". Однако, если в качестве параметра cv указать cv=[(slice(None), slice(None))], тогда никакого перемешивания не будет, и мы сможем подбирать параметры через GridSearchCV в комбинации с Pipeline.

Однако, можно сделать еще проще и использовать TimeSeriesSplit, чем мы и воспользуемся.

```python
def model_plot(y_test, y_pred):
    """Функция строит графики для оценки модели.
    Аргументы:
    y_test - тестовые данные целевого признака;
    y_pred - предсказанные данные целевого признака"""

    plt.figure(figsize=(15, 7))
    plt.plot(y_pred, "r", label="Предсказание")
    plt.plot(y_test.values, label="Действительные значения")
    plt.legend(loc="best")
    plt.title("Визуализация предсказаний модели")
    plt.grid(True)
```

```python
def model_score_test(X_train, X_test, y_train, y_test, name, model, grid_search=False):
    """Функция строит предсказания модели,  может подбирать наилучшие
    варианты преобразований и параметров с помощью Pipeline и GridSearchCV.
    Результатом работы функции являются график оценки модели,
    вывод наилучших вариантов преобразования модели,
    таблица со значениями метрик RMSE, R2.
    Аргументы:
    X_train - тренировочные данные для обучения модели;
    X_test - тестовые данные для предсказания целевого признака;
    y_train - значения целевого признака для обучения модели;
    y_test - значение целевого признака для вычисления метрик;
    name - наименование объекта;
    model - модель;
    param_grid - сетка параметров."""

    global comparison_of_models
    if grid_search:
        start = timer()
        cv = TimeSeriesSplit(n_splits=5)
        preprocessor = Pipeline(
            steps=[("scaler", "passthrough"), ("feature_selection", "passthrough")]
        )
        regressor = Pipeline(steps=[("preprocessor", preprocessor), ("model", model),])

        grid_search = GridSearchCV(
            regressor,
            param_grid=grid_search,
            cv=cv,
            scoring="neg_root_mean_squared_error",
            n_jobs=-1,
        )
        grid_search.fit(X_train, y_train)
        end = timer()
        fit_time = str(timedelta(seconds=end - start))

        best_params = grid_search.best_params_

        start = timer()
        y_pred = grid_search.predict(X_test)
        end = timer()
        predict_time = str(timedelta(seconds=end - start))
    else:
        start = timer()

        model.fit(X_train, y_train)
        end = timer()
        fit_time = str(timedelta(seconds=end - start))

        start = timer()
        y_pred = model.predict(X_test)
        end = timer()
        predict_time = str(timedelta(seconds=end - start))

    plot = model_plot(y_test, y_pred)
    plot

    pred_MSE = mean_squared_error(y_test, y_pred)
    pred_RMSE = sqrt(pred_MSE)
    pred_R2 = r2_score(y_test, y_pred)

    temp_data = pd.DataFrame(
        {
            "Наименование алгоритма": [name],
            "Время обучения (сек.)": [fit_time],
            "Скорость предсказания (сек.)": [predict_time],
            "RMSE на обучении": [abs(grid_search.best_score_)],
            "RMSE предсказания": [pred_RMSE],
            "R2 предсказания": [pred_R2],
        }
    )
    if len(comparison_of_models) == 0:
        comparison_of_models = comparison_of_models.append(temp_data)
    elif name not in list(comparison_of_models["Наименование алгоритма"]):
        comparison_of_models = comparison_of_models.append(temp_data)
    else:
        comparison_of_models[
            comparison_of_models["Наименование алгоритма"] == name
        ] = temp_data

    comparison_of_models = comparison_of_models.reset_index(drop=True)
    if grid_search:
        print(
            f"Для улучшения модели использовались следующие параметры: {best_params}",
            end="\n\n",
        )
    display(comparison_of_models)
    return comparison_of_models, plot
```

```python
param_grid_lin_regr = [
    {"preprocessor__scaler": [None, RobustScaler(), StandardScaler()]}
]
```

```python
comparison_of_models, graph_lin_regr_default = model_score_test(
    X_train_default,
    X_test_default,
    y_train_default,
    y_test_default,
    "LinearRegression_default",
    LinearRegression(),
    param_grid_lin_regr,
)
```

```python
comparison_of_models, graph_lin_regr_boxcox = model_score_test(
    X_train_boxcox,
    X_test_boxcox,
    y_train_boxcox,
    y_test_boxcox,
    "LinearRegression_boxcox",
    LinearRegression(),
    param_grid_lin_regr,
)
```

Мы видим, что значение метрики RMSE в значительной степени упало, когда для обучения мы использовали данные с преобразованием Бокса-Кокса. Может показаться, что такое малое значение RMSE получилось за счет лика данных. Но на самом деле, на 2 и 3 графиках по оси У видно, что значения числа заказов такси за счет преобразований в значительной степени уменьшились, поэтому для объективной интерпретации точности модели стоит обратить внимание на значение метрики R2.
