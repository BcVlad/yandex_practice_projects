### Обработка пропусков

Нам предстоит выяснить, в результате чего появились пропущенные значения в каждом из столбцов, после чего для каждого конкретного случая выберем способ обработки пропущенных значений

Сначала проверим наш датафрэйм на наличие дубликатов. Если они есть, избавимся от них

```python
estate_df.duplicated().sum()
```

Вновь посмотрим, в каких переменных у нас имеются пропущенные значения

Тут важно отметить, что по каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма.

Получается, что часть пропущенных значений (например, ceiling_height, floors_total) появились в результате того, что эти сведения не занёс пользователь, а другая часть (например, airports_nearest, cityCenters_nearest, parks_around3000, parks_nearest, ponds_around3000, ponds_nearest, days_exposition) - ввиду несовершенной работы алгоритмов, которые автоматически должны вычеслять такие параметры, как наличие парков рядом, а также расстояние до центра города.

В первом и во втором случае пропуски не являются случайными, и механизм их появления можно отнести к MNAR (Missing Not At Random) — механизм формирования пропусков, при котором данные отсутствуют в зависимости от неизвестных факторов. MNAR предполагает, что вероятность пропуска могла бы быть описана на основе других атрибутов, но информация по этим атрибутам в наборе данных отсутствует.

В столбцах ceiling_height, floors_total, locality_name пропуски появились из-за того, что пользователь, разместивший объявление, не указал высоту потолков в квартире, этажность дома, а также населенный пункт. Уместнее всего в данном случае не выполнять замен пропущенных значений. Замена на среднее или медианное значение в данном случае может существенно исказить распределение выборки.

В таких столбцах как airports_nearest, cityCenters_nearest, parks_around3000, parks_nearest, ponds_around3000, ponds_nearest, days_exposition, как было описано выше, пропуски появились из-за особенностей работы алгоритмов по автоматическому заполнению соответствующих значений. Их также не будем заполнять: часть из этих переменных является номинативными, часть - количественными, и подобрать какой-то алгоритм для замены пропущенных значений не представляется возможным.


Отсутствие значений — тоже важный сигнал, от которого не всегда имеет смысл избавляться.

Приведем столбцы, пропуски в которых мы можем обработать: living_area, is_apartment, kitchen_area, balcony. От пропусков в living area и kitchen area можно попробовать избавиться, выдвинув гипотезу, что данные переменные сильно коррелируют с переменной total_area, в которой у нас пропусков нет. Для того, чтобы проверить эту гипотезу, построим матрицу корреляций  

```python
estate_df.corr()
```

Гипотеза подтвердилась: корреляция между total_area и living_area составляет 0,94. Корреляция между total_area и kitchen_area ниже: 0,61. Попробуем восстановить пропущенные значения с помощью линейной регрессии. В качестве 'x' будут использоваться значения total_area, а в качестве 'y' - значения living_area. Важно 'x' представить в качестве двумерного массива.

```python
X = estate_df.query('~living_area.isnull()')[['total_area']].values
y = estate_df.query('~living_area.isnull()')['living_area'].values
```

```python
model = LinearRegression().fit(X, y)
```

Убедимся в том, что мы получили приемлимые результаты. Критерием корректности построения модели и применения линейной регрессии может служить коэффициент детерменации (R^2 — R-квадрат) — это доля дисперсии зависимой переменной, объясняемая рассматриваемой моделью зависимости.

```python
r_sq = model.score(X, y)
print('coefficient of determination:', r_sq)
```

Получаем, что 88% изменчивости зависимой переменной (у) обуславливается взаимосвязью с независимой переменной (х). Это хороший результат, и применение ленейной регрессии в данном случае целесообразно

```python
# Предскажем пропущенные значения переменной living_area

y_pred = model.predict(estate_df.query('living_area.isnull()')[['total_area']].values)
```

```python
estate_df.loc[estate_df.living_area.isnull(), 'living_area'] = y_pred
```

Проверим корректность преобразований

```python
len(estate_df.loc[estate_df.living_area.isnull(), 'living_area'])
```

```python
# На всякий случай выполним проверку на то, что отсутствуют наблюдения, где значение переменной total_area
# строго меньше значения переменной living_area

len(estate_df.query('total_area < living_area'))
```

Округлим значения переменной 'living_area' до 2 знаков после запятой.

```python
estate_df.living_area = estate_df.living_area.apply(lambda x: round(x, 2))
```

Также используем линейную регрессию, чтобы сделать предсказание значений переменной kitchen_area

```python
X = estate_df.query('~kitchen_area.isnull()')[['total_area']].values
y = estate_df.query('~kitchen_area.isnull()')['kitchen_area'].values
```

```python
model = LinearRegression().fit(X, y)
```

```python
r_sq = model.score(X, y)
print('coefficient of determination:', r_sq)
```

Коэффициент детерминации оказался существенно ниже, чем в случае с переменной living_area

```python
# Предскажем пропущенные значения переменной kitchen_area

y_pred = model.predict(estate_df.query('kitchen_area.isnull()')[['total_area']].values)
```

```python
estate_df.loc[estate_df.kitchen_area.isnull(), 'kitchen_area'] = y_pred
```

Проверим корректность преобразований

```python
len(estate_df.loc[estate_df.kitchen_area.isnull(), 'kitchen_area'])
```

```python
# На всякий случай выполним проверку на то, что отсутствуют наблюдения, где значение переменной total_area
# строго меньше значения переменной kitchen_area

len(estate_df.query('total_area < kitchen_area'))
```

Округлим значения переменной 'kitchen_area' до 2 знаков после запятой.

```python
estate_df.kitchen_area = estate_df.kitchen_area.apply(lambda x: round(x, 2))
```

Построим модели для living_area и kitchen_area, которые основаны на столбце с отношением living_area или kitchen_area к total_area. Для этого добавим соответствующие столбцы в нашу таблицу.

```python
model_df = estate_df
```

```python
model_df['living_prop'] = model_df['living_area'] / model_df['total_area']
model_df['kitchen_prop'] = model_df['kitchen_area'] / model_df['total_area']
```

```python
X = model_df.query('~living_area.isnull()')[['living_prop']].values
y = model_df.query('~living_area.isnull()')['living_area'].values
model = LinearRegression().fit(X, y)
r_sq = model.score(X, y)
print('coefficient of determination:', r_sq)
```

```python
X = model_df.query('~kitchen_area.isnull()')[['kitchen_prop']].values
y = model_df.query('~kitchen_area.isnull()')['kitchen_area'].values
model = LinearRegression().fit(X, y)
r_sq = model.score(X, y)
print('coefficient of determination:', r_sq)
```

Значения R^2 получились меньше, всего 16% изменчивости зависимой переменной (living_area) обуславливается взаимосвязью с независимой переменной (living_prop). И 14% изменчивости зависимой переменной (kitchen_area) обуславливается взаимосвязью с независимой переменной (kitchen_prop).

Таким образом, оставляем рабочим вариантом предыдущую версию модели, основанную на взаимосвязи с переменной total_area.

### Корректировка типов данных

Проанализируем и приведем данные к нужным типам данных

Создадим функцию, чтобы преобразовать значения тех колонок, которые мы зададим из float в int. Преобразуем значения следующих столбцов:

* 'last_price'
* 'floors_total'
* 'balcony'
* 'airports_nearest'
* 'cityCenters_nearest'
* 'parks_around3000'
* 'parks_nearest'
* 'ponds_around3000'
* 'ponds_nearest'
* 'days_exposition'

Столбцы 'total_area', 'ceiling_height', 'living_area', 'kitchen_area' преобразовывать не будем, в них приемлемы значения с типом данных float.

```python
def float_to_int(df, columns):
    for column in columns:
        try:
            df.loc[:, column] = df.loc[:, column].astype('int')
        except:
            try:
                df.loc[:, column] = df.loc[:, column].astype('Int64')
            except:
                df.loc[:, column] = pd.to_numeric(df.loc[:, column].round(), errors='coerce').astype('Int64')
```

```python
columns = ['last_price', 
           'floors_total', 
           'balcony', 
           'airports_nearest', 
           'cityCenters_nearest', 
           'parks_around3000', 
           'parks_nearest',
           'ponds_around3000',
           'ponds_nearest',
           'days_exposition']

float_to_int(estate_df, columns)
```

```python
estate_df.info()
```

Преобразования прошли успешно

Напишем функции для автоматизации построения графиков.

```python
def boxplot(dataframe, clmn, without_outliers=False, rng=None):
    """Функция автоматизирует построение графика boxplot. 
    Для наглядности при without_outliers=True
    взяли не 1,5 межквартильных размаха, а 2"""
    if not rng and without_outliers:
        Q1 = dataframe[clmn].quantile(0.25) 
        Q3 = dataframe[clmn].quantile(0.75)
        IQR = Q3 - Q1
        a = Q1 - 2 * IQR
        b = Q3 + 2 * IQR
        rng = (a, b)
        
    dataframe.boxplot(column=clmn, grid=True)
    
    name = clmn
    try:
        name = name.split('_')
        name = ' '.join(name)
        plt.ylabel(f"{name} values")
    except:
        plt.ylabel(f"{name} values")
    plt.ylim(rng)
    if without_outliers:
        plt.title(f"Boxplot of {name} without outliers")
    else:
        plt.title(f"Boxplot of {name}")
    
    plt.show()
```

```python
def hist(dataframe, clmn, bns, without_outliers=False, rng=None):
    """Функция автоматизирует построение графика hist. 
    Для наглядности при without_outliers=True
    взяли не 1,5 межквартильных размаха, а 2"""
    if not rng and without_outliers:
        Q1 = dataframe[clmn].quantile(0.25) 
        Q3 = dataframe[clmn].quantile(0.75)
        IQR = Q3 - Q1
        a = max(Q1 - 2 * IQR, dataframe[clmn].min())
        b = Q3 + 2 * IQR
        rng = (a, b)
            
    dataframe.plot(kind='hist',
                   y=clmn,
                   bins=bns,
                   legend=True,
                   range = rng,
                   grid=True)
    name = clmn
    try:
        name = name.split('_')
        name = ' '.join(name)
        plt.xlabel(f"{name} values")
    except:
        plt.xlabel(f"{name} values")
    if without_outliers:
        plt.title(f"Histogram of {name} without outliers")
    else:
        plt.title(f"Histogram of {name}")
    
    plt.show()
```

```python
def vizualization_with_without_outliers(dataframe, clmn, bns_with_o=10, bns_without_o=10, rage_for_boxplot=None, rage_for_hist=None, number_of_graph = 5):
    """Вывод нескольких графиков с учетом выбросов и без них"""
    if number_of_graph == 5:
        boxplot(dataframe, clmn, without_outliers=False, rng=rage_for_boxplot)
        boxplot(dataframe, clmn, without_outliers=True)
        hist(dataframe, clmn, bns_with_o, without_outliers=False, rng=rage_for_hist)
        hist(dataframe, clmn, bns_without_o, without_outliers=True)
    elif number_of_graph == 1:
        boxplot(dataframe, clmn, without_outliers=False, rng=rage_for_boxplot)
    elif number_of_graph == 2:
        boxplot(dataframe, clmn, without_outliers=True)
    elif number_of_graph == 3:
        hist(dataframe, clmn, bns_with_o, without_outliers=False, rng=rage_for_hist)
    elif number_of_graph == 4:
        hist(dataframe, clmn, bns_without_o, without_outliers=True)
```

### Анализ редких и выбивающихся значений

Ранее мы выявляли аномальные значения в наших данных. Попробуем найти новые аномальные значения, а также обработать их.

Существуют разные способы идентификации аномальных и выбивающихся значений. Например, существует правило 2 и 3 сигм: правило, утверждающее, что вероятность того, что случайная величина отклонится от своего математического ожидания более чем на три среднеквадратических отклонения, практически равна нулю. К сожалению, для его применения нужно быть уверенным, что переменная распределена нормально, в противном случае пользоваться им нельзя. Нам нужен какой-то универсальный алгоритм для идентификации значений, которые выбиваются из общей выборки.

Для удобства обработки аномальных значений напишем функцию, которая автоматизирует процесс выявления редких и выбивающихся значений через межквартильное расстояние. По умолчанию функция будет отсеивать значения, которые превышают "усы" боксплота (длина "усов" по умолчанию 1,5 межквартильных расстояний). В случае, если потеря данных составляет меньше или равно 2,5%, функция обработает их автоматически.

```python
def abnormal_outliers_val(dataframe, clmn, confidence_interval=1.5):
    """Функция принимает датафрэйм и анализируемую колонку.
    Отсекает значения, которые ниже или выше значений усов боксплота"""
    length = len(dataframe)
    new_dataframe = dataframe.copy()
    Q1 = dataframe[clmn].quantile(0.25) 
    Q3 = dataframe[clmn].quantile(0.75)
    IQR = Q3 - Q1
    new_dataframe = new_dataframe.loc[((new_dataframe[clmn] >= (Q1 - confidence_interval * IQR)) & (new_dataframe[clmn] <= (Q3 + confidence_interval * IQR))) | (new_dataframe[clmn].isnull())]
    new_length = len(new_dataframe)
    data_loss = 1 - (new_length / length)
    if data_loss == 0:
        print(f'Потеря данных составляет {data_loss}, данные не изменились')
        return dataframe
    elif data_loss <= 0.025:
        print(f'Потеря данных составляет {data_loss}, выбросы и аномальные значения устранены автоматически. Количество отсеянных наблюдений {length - new_length} шт.')
        return new_dataframe
    else:
        print(f'Потеря данных составляет {data_loss} ({length - new_length}  шт.), рекомендуется обработать данные вручную или увеличить межквартильный интервал (по умолчанию 1.5), чтобы избежать чрезмерных потерь данных')
        return dataframe
```

Опишем взаимосвязь математически. Регрессионная прямая описывается формулой y = b0 + b1 * x, где b0 (intercept) - показывает, где прямая пересекает ось Y. А b1 (slope) - угол наклона регрессионной прямой. Найдем значения b0 и b1, а затем интерпретируем их.

```python
X = clean_df_iter6[['total_area']].values
y = clean_df_iter6['last_price'].values
```

```python
slr = LinearRegression()
slr.fit(X, y)
y_pred = slr.predict(X)
print('Slope: {:.2f}'.format(slr.coef_[0]))
print('Intercept: {:.2f}'.format(slr.intercept_))
```

Мы получили, что при увеличении площади на 1 м2, стоимость недвижимости увеличивается в среднем на 141 634 руб.
