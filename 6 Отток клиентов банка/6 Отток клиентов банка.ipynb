```python
def load_df(name_of_df, separator=",", ind=None):
    """Функция загружает датафрэйм и выводит основные данные по нему.
    Аргументы:
    name_of_df - путь к датафрэйму;
    separator - разделитель;
    ind - столбцы для использования в качестве номеров строк"""

    df = pd.read_csv(name_of_df, sep=separator, index_col=ind)
    print("Выведем первые 5 строк таблицы:")
    display(df.head())
    print()
    print("Отобразим основные описательные статистики по признакам:")
    display(df.describe())
    print("Получим общую информацию о таблице:")
    display(df.info())
    print()
    print(f"Пропущенные значения по признакам:")
    print(df.isnull().sum())
    return df
```

Создадим пустой датафрэйм, куда будем записывать значения F1-меры, score, а также площади под ROC-кривой модели. Датафрэйм будет заполнятьс автоматически в теле функции, а сортировка будет выполнена по убыванию значений F1-меры.

```python
def roc(clf, train_data, test_data, i):
    """Функция выводит график ROC-кривой.
    Аргументы:
    clf - классификатор;
    train_data - список с данными для обучения модели;
    test_data - список с данными для тестирования модели;
    i - номер итерации в теле функции get_score"""

    y_predicted_prob = clf.predict_proba(test_data[i][0])
    fpr, tpr, thresholds = roc_curve(test_data[i][1], y_predicted_prob[:, 1])
    roc_auc = auc(fpr, tpr)
    lw = 2
    plt.figure()
    plt.plot(
        fpr, tpr, color="darkorange", lw=lw, label="ROC curve (area = %0.5f)" % roc_auc
    )
    plt.plot([0, 1], [0, 1], color="navy", lw=lw, linestyle="--")
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curve")
    plt.legend(loc="lower right")
    return (roc_auc, plt.show())
```

```python
def get_score(
    model,
    params,
    train_data,
    test_data,
    n_iter,
    tree_algoritm=None,
    xg_boost=None,
    num_round=None,
    cv=5,
    upsampled=False,
):
    """Функция выводит score модели, а также
    значение F1-метрики.
    Аргументы:
    model - модель;
    params - параметры модели;
    train_data - список с данными для обучения модели;
    test_data - список с данными для тестирования модели;
    n_iter - число итераций при подборе параметров;
    tree_algoritm - базируется ли алгоритм на деревьях решений;
    cv - параметр кросс-валидации"""

    global comparison_of_models
    best_score = [0, 0, 0]
    transformation = {1: "standartization", 2: "normalization", 3: "robust_scaler"}
    if tree_algoritm:
        clf = RandomizedSearchCV(
            model,
            param_distributions=params,
            n_iter=n_iter,
            cv=cv,
            verbose=0,
            random_state=12345,
            n_jobs=-1,
        )
        clf.fit(train_data[0][0], train_data[0][1])
        score = clf.score(test_data[0][0], test_data[0][1])
        y_pred = clf.predict(test_data[0][0])
        f1 = f1_score(test_data[0][1], y_pred, average="macro")
        print(f"Значение метрики score: {score}")
        print(f"Значение F1-меры: {f1}")
        print(
            f"Значения метрик обеспечиваются при следующих параметрах модели: {clf.best_params_}",
            end="\n\n",
        )
        rc = roc(clf, train_data, test_data, 0)
        rc[1]
        best_score[0], best_score[1], best_score[2] = f1, score, rc[0]
        mdl = str(clf.best_estimator_).split("(")[0]
        
    else:
        for i, element in enumerate(train_data):
            clf = RandomizedSearchCV(
                model,
                param_distributions=params,
                n_iter=n_iter,
                cv=cv,
                verbose=0,
                random_state=12345,
                n_jobs=-1,
            )
            clf.fit(train_data[i][0], train_data[i][1])
            score = clf.score(test_data[i][0], test_data[i][1])
            y_pred = clf.predict(test_data[i][0])
            f1 = f1_score(test_data[i][1], y_pred, average="macro")
            print(f"Значение метрики score: {score}")
            print(f"Значение F1-меры: {f1}")
            print(
                f"Значения метрик обеспечиваются при следующих параметрах модели: {clf.best_params_}",
                end="\n\n",
            )
            rc = roc(clf, train_data, test_data, i)
            rc[1]
            if f1 > best_score[0]:
                best_score[0], best_score[1], best_score[2] = f1, score, rc[0]
                if i > 0:
                    mdl = "_".join([str(clf.best_estimator_).split("(")[0], transformation[i]])
                else:
                    mdl = str(clf.best_estimator_).split("(")[0]
    
    if upsampled:
        mdl = "_".join([mdl, 'upsampled'])
    temp_data = pd.DataFrame(
        {
            "model": [mdl],
            "f1_score": [best_score[0]],
            "score": [best_score[1]],
            "auc_roc": [best_score[2]],
        }
    )
    comparison_of_models = comparison_of_models.append(temp_data)
    comparison_of_models = comparison_of_models.sort_values("f1_score", ascending=False)
    comparison_of_models = comparison_of_models.reset_index(drop=True)
    display(comparison_of_models)
    return clf.best_estimator_
```

## Борьба с дисбалансом

Для борьбы с дисбалансом классов используются методы upsampling - увеличение выборки, и downsampling - уменьшение выборки.
Преобразование upsampling проходит в несколько этапов:
* Разделить обучающую выборку на отрицательные и положительные объекты;
* Скопировать несколько раз положительные объекты;
* С учётом полученных данных создать новую обучающую выборку;
* Перемешать данные: идущие друг за другом одинаковые вопросы не помогут обучению.

Downsampling происходит по схожему сценарию:
* Разделить обучающую выборку на отрицательные и положительные объекты;
* Случайным образом отбросить часть из отрицательных объектов;
* С учётом полученных данных создать новую обучающую выборку;
* Перемешать данные. Положительные не должны идти следом за отрицательными: алгоритмам будет сложнее обучаться.

Исправим баланс классов с помощью upsampling. В использовании Downsampling в данном случае нет особого смысла: его стоит использовать, когда имеется огромное число объектов. В данном случае уменьшение данных неминуемо приведет к снижению показателей основных метрик для задачи классификации.

```python
def upsample(X, y, repeat):
    """Функция принимает на вход датасэты с обучающими 
    признаками и целевым признаком. Затем увеличивает 
    количество объектов меньшего класса.
    Аргументы:
    X - датасэт с признаками;
    y - целевой признак"""

    features_zeros = X[y == 0]
    features_ones = X[y == 1]
    target_zeros = y[y == 0]
    target_ones = y[y == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)

    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345
    )

    return features_upsampled, target_upsampled
```

Проведем финальное тестирование модели VotingClassifier и отразим ее метрики.

Прежде чем мы рассчитаем значения основных метрик модели, поясним, что они означают. В общем случае при решении задач классификации имеет смысл рассчитывать следующие метрики:

* Accuracy (только при сбалансированных выборках)
* Precision (точность)
* Recall (полнота)
* F1-мера
* AUC-ROC

Precision можно интерпретировать как долю объектов, названных классификатором положительными и при этом действительно являющимися положительными, а recall показывает, какую долю объектов положительного класса из всех объектов положительного класса нашел алгоритм.

Именно введение precision не позволяет нам записывать все объекты в один класс, так как в этом случае мы получаем рост уровня False Positive. Recall демонстрирует способность алгоритма обнаруживать данный класс вообще, а precision — способность отличать этот класс от других классов.

Precision и recall не зависят, в отличие от accuracy, от соотношения классов и потому применимы в условиях несбалансированных выборок.

F1-мера (в общем случае) — среднее гармоническое precision и recall.


Кроме того, одним из способов оценить модель в целом, не привязываясь к конкретному порогу, является AUC-ROC (или ROC AUC) — площадь (Area Under Curve) под кривой ошибок (Receiver Operating Characteristic curve ). Данная кривая представляет из себя линию от (0,0) до (1,1) в координатах True Positive Rate (TPR) и False Positive Rate (FPR).
